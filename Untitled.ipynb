```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from skimage import data
from skimage.filters import threshold_otsu, threshold_adaptive, threshold_isodata, threshold_li, threshold_yen, denoise_bilateral, denoise_tv_bregman, denoise_tv_chambolle 
from skimage.segmentation import clear_border
from skimage.measure import label
from skimage.morphology import closing, square
from skimage.measure import regionprops
from skimage.color import label2rgb
from skimage.data import imread
from skimage.io import imsave
import sys
import os
```

```python
def apply_threshold(image, threshold, **params):
    thresh = globals()[threshold](image, **params)
    if thresh.shape != image.shape:
#         print("Threshold", thresh)
        res = closing(image > thresh, square(3))
    else: 
        res = thresh
    return res
```

```python
def threshold_percentile(image, percentile=[2, 90, 96]):   
    p2, p90, p96 = np.percentile(image.flatten(), percentile)
    bw = (image < p2) | ((image < p96) & (image > p90))
    return bw

def detect_whale(file_in, file_out, method):
    image = imread(file_in, as_grey=True)
    # apply threshold
    params=dict()
    params["threshold_adaptive"] = dict(block_size=201)
    params["denoise_tv_bregman"] = dict(weight=10)
    
    bw = apply_threshold(image, method, **params.get(method, dict()))

    # remove artifacts connected to image border
    cleared = bw.copy()
    clear_border(cleared)

    # label image regions
    label_image = label(cleared)

    m = 0
    for region in regionprops(label_image):
        if region.area > m:
            m = region.area
            minr, minc, maxr, maxc = region.bbox
    
    image_rgb = imread(file_in)

    image_rgb[minr-5:minr, minc-5:maxc+5] = (255, 0, 0)
    image_rgb[maxr:maxr+5, minc-5:maxc+5] = (255, 0, 0)
    image_rgb[minr:maxr, maxc:maxc+5] = (255, 0, 0)
    image_rgb[minr:maxr, minc-5:minc] = (255, 0, 0)
    imsave(file_out, image_rgb)
```
